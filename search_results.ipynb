```python
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
pytrends = TrendReq(hl='en-US', tz=360)

# Define your keywords
keywords = ['Burberry', 'CHANEL']

# Build payload
pytrends.build_payload(kw_list=keywords)

# Get interest over time
interest_over_time_df = pytrends.interest_over_time()

print(interest_over_time_df)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(interest_over_time_df['Burberry'], label='Burberry', marker='o')
plt.plot(interest_over_time_df['CHANEL'], label='CHANEL', marker='o')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Interest Over Time')
plt.title('Comparison of Burberry and Chanel Popularity over Time')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
